```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

## Preparing the variables

```python
boston_X, boston_y = load_boston(return_X_y=True)
boston_X_train, boston_X_test, boston_y_train, boston_y_test = train_test_split(boston_X, boston_y)
```

## Linear Regression

```python
regr = linear_model.LinearRegression()
regr.fit(boston_X_train, boston_y_train)

boston_y_pred = regr.predict(boston_X_test)

# print(f'Coefficients: {regr.coef_}')
print(f'Mean squared error: {mean_squared_error(boston_y_test, boston_y_pred):.2f}')
print(f'Coefficient of determination: {r2_score(boston_y_test, boston_y_pred):.2f}')
plt.scatter(boston_y_test, boston_y_pred)
plt.show()
```
